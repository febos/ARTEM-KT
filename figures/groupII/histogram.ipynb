```python
import matplotlib.pyplot as plt

```

```python
dat = '''ARTEM12	4	0	3	1.0	0.57	0.73
ARTEM11	6	8	1	0.43	0.86	0.57
RNAMotifScanX	3	0	4	1.0	0.43	0.6
DSSR	2	1	5	0.67	0.29	0.4
FR3D	0	0	7	0	0	0
JAR3D	0	0	7	0	0	0
LocalSTAR3D	0	0	7	0	0	0
CircularSTAR3D	0	0	7	0	0	0'''.split('\n')[::-1]

dat = [x.split('\t') for x in dat]

names = [x[0] for x in dat]
prec  = [float(x[-3]) for x in dat]
rec   = [float(x[-2]) for x in dat]
fs    = [float(x[-1]) for x in dat]
names
```

```python
labels = ['ARTEM\n≥12nt',
 'ARTEM\n≥11nt',
 'RNAMotifScanX',
 'DSSR',
 'FR3D',
 'JAR3D',
 'LocalSTAR3D',
 'CircularSTAR3D'][::-1]

fig = plt.figure(
    figsize=(7, 7),  # figure size in inches
)

bar3 = plt.barh([x+0.3 for x in range(len(labels))][4:], fs[4:], height=0.3, label = "F-score", color='black')
plt.bar_label(bar3, padding =5, fontsize=14)
bar2 = plt.barh([x for x in range(len(labels))][4:], rec[4:], height=0.3, label = "Recall", color='gray')
plt.bar_label(bar2, padding =5, fontsize=14)
bar1 = plt.barh([x-0.3 for x in range(len(labels))][4:], prec[4:], height=0.3, label = "Precision", color='lightgrey')
plt.bar_label(bar1, padding =5, fontsize=14)
plt.yticks(range(len(labels))[4:],labels=labels[4:], fontsize=16)
plt.xticks([0.0,0.5,1.0,1.2],labels=[0.0,0.5,1.0,''],fontsize=14)
plt.legend(loc="lower right", fontsize=14)
plt.savefig('Fig8C.svg')
plt.show()
```
